## Handling missing values using - COMPLETE CASE ANALYSIS

```python
data.isnull().mean()*100
```

```python
cols = [var for var in data.columns if data[var].isnull().mean() < 0.05 and data[var].isnull().mean() > 0]
```

```python
cols
```

```python
len(data[cols].dropna())/len(data)
```

```python
new_df = data[cols].dropna()
```

```python
data.shape,new_df.shape
```

```python
new_df.hist(bins=50,density=True,figsize=(12,12))
plt.show()
```

```python
fig = plt.figure()
ax = fig.add_subplot(111)
data['1st_Road_Number'].hist(bins=50,density=True,ax=ax,color='red')
new_df['1st_Road_Number'].hist(bins=50,ax=ax,density=True,color='blue',alpha=0.8)
plt.show()
```

```python
fig = plt.figure()
ax = fig.add_subplot(111)
data['2nd_Road_Number'].plot.density(color='red')
new_df['2nd_Road_Number'].plot.density(color='blue')
```

```python
temp = pd.concat([
    data['Light_Conditions'].value_counts() / len(data),
    new_df['Light_Conditions'].value_counts()/len(new_df)
],axis=1)
temp.columns = ['originals','cca']
temp
```

## Percentage of Null values

```python
null_columns = pd.DataFrame(data[data.columns[data.isnull().any()]].isnull().sum() * 100 / data.shape[0],columns=['Percentage-of-NAN-values'])
```

```python
null_columns['Total-NAN-values'] = data[data.columns[data.isnull().any()]].isnull().sum()
```
